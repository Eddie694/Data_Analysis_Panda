```python
group_three = df[['number_bedrooms', 'price']]
df_group_three = group_three.groupby(['number_bedrooms'], as_index=False).sum()
df_group_three.head()
```

```python
sns.boxplot(x='number_bedrooms', y = 'price', data=df_group_three)
```

```python
plt.pcolor(df_group_three, cmap='RdBu')
plt.colorbar()
plt.show()
```

```python
fig, ax = plt.subplots()
im = ax.pcolor(df_group_three, cmap='RdBu')

#label names
row_labels = df_group_three.columns.levels[1]
col_labels = df_group_three.index

#move ticks and labels to the center
ax.set_xticks(np.arange(df_group_three.shape[0]) + 0.5, minor=False)
ax.set_yticks(np.arange(df_group_three.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```
